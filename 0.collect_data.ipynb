### data collection

```python
wav_data_folder = Path('./data')
mp3_data_folder = Path('./mp3data')
```

```python
sounds_meta_file = Path('soundmeta.csv')
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
temp_data = []
for idx,audio_url in enumerate(audio_file_list.iloc[:,0]):
    resp = requests.get(audio_url)
    soup = BeautifulSoup(resp.text, 'html.parser')
    contents = soup.find('h1',itemprop='name').find_all('a')
    track_name = contents[0].text
    author = contents[1].text
    genre_tag = soup.find('meta',itemprop='genre')
    genre=None
    if genre_tag:
        genre = genre_tag['content']
    play_count = soup.find('meta',property='soundcloud:play_count')['content']
    download_count = soup.find('meta',property='soundcloud:download_count')['content']
    comments_count = soup.find('meta',property='soundcloud:comments_count')['content']
    like_count = soup.find('meta',property='soundcloud:like_count')['content']
    temp_data.append([idx,track_name,author,genre,play_count,download_count,comments_count,like_count,audio_url])
    print(f'finish idx {idx} of {len(audio_file_list.index)-1}')
```

```python
df = pd.DataFrame(temp_data, columns=['index', 'track_name', 'author','genre','play_count','download_count','comments_count','like_count','audio_url'])
```

```python
df
```

```python
df.to_csv(str(sounds_meta_file),index=False)
```

```python
from soundwave_api import SoundcloudAPI,Track
```

```python
api = SoundcloudAPI()
```

```python
df
```

```python
for idx,name,url in zip(df.index,df.track_name,df.audio_url):
    if idx<=37: continue
    track = api.resolve(url)
    assert type(track) is Track
    filename = wav_data_folder / f'{idx}.mp3'
    with open(str(filename), 'wb+') as fp:
        track.write_mp3_to(fp)
    print(f'finished {idx}')
```
